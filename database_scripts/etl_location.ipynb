# Upload to Database

```python
# Read database.ini
filename='../database.ini'
section='postgresql'
parser = ConfigParser()
parser.read(filename)
config = {}
if parser.has_section(section):
    params = parser.items(section)
    for param in params:
        config[param[0]] = param[1]
else:
    raise Exception(f'Section {section} not found in the {filename} file')
```

```python
try:
    # Create an SQLAlchemy engine
    db = create_engine(f"postgresql+psycopg2://{config['user']}:{config['password']}@{config['host']}:{config['port']}/{config['database']}")
    conn = db.connect()

    # Save the DataFrame to the table
    dfCombined.to_sql('location', con=conn, if_exists='append', index=False) 

    # Commits the changes and closes the connection
    conn.commit()
    conn.close()
    print("Database Upload Complete")
except Exception as e:
    print(f"Error: {e}")
```

```python
# Queries the database to check results
conn = psycopg2.connect(**config) 
cursor = conn.cursor() 
sql1 = '''select * from location;'''
cursor.execute(sql1)
for i in cursor.fetchall():
    print(i) 
conn.close() 
```
