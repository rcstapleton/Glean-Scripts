# Data Load and Preparation

```python
# Load data into dataframes
chr_2010 = pd.read_csv(r"../data/CHR/analytic_data2010.csv", low_memory=False)
chr_2011 = pd.read_csv(r"../data/CHR/analytic_data2011.csv", low_memory=False)
chr_2012 = pd.read_csv(r"../data/CHR/analytic_data2012.csv", low_memory=False)
chr_2013 = pd.read_csv(r"../data/CHR/analytic_data2013.csv", low_memory=False)
chr_2014 = pd.read_csv(r"../data/CHR/analytic_data2014.csv", low_memory=False)
chr_2015 = pd.read_csv(r"../data/CHR/analytic_data2015.csv", low_memory=False)
chr_2016 = pd.read_csv(r"../data/CHR/analytic_data2016.csv", low_memory=False)
chr_2017 = pd.read_csv(r"../data/CHR/analytic_data2017.csv", low_memory=False)
chr_2018 = pd.read_csv(r"../data/CHR/analytic_data2018_0.csv", low_memory=False)
chr_2019 = pd.read_csv(r"../data/CHR/analytic_data2019.csv", low_memory=False)
chr_2020 = pd.read_csv(r"../data/CHR/analytic_data2020_0.csv", low_memory=False)
chr_2021 = pd.read_csv(r"../data/CHR/analytic_data2021.csv", low_memory=False)
chr_2022 = pd.read_csv(r"../data/CHR/analytic_data2022.csv", low_memory=False)
chr_2023 = pd.read_csv(r"../data/CHR/analytic_data2023_0.csv", low_memory=False)
print("Loadng Complete")
```

```python
# Drop the unneeded 2nd header row
chr_2010 = chr_2010.drop(index=0)
chr_2011 = chr_2011.drop(index=0)
chr_2012 = chr_2012.drop(index=0)
chr_2013 = chr_2013.drop(index=0)
chr_2014 = chr_2014.drop(index=0)
chr_2015 = chr_2015.drop(index=0)
chr_2016 = chr_2016.drop(index=0)
chr_2017 = chr_2017.drop(index=0)
chr_2018 = chr_2018.drop(index=0)
chr_2019 = chr_2019.drop(index=0)
chr_2020 = chr_2020.drop(index=0)
chr_2021 = chr_2021.drop(index=0)
chr_2022 = chr_2022.drop(index=0)
chr_2023 = chr_2023.drop(index=0)
print("2nd Row Dropped")


# Reset the index after removing the unneeded 2nd header row
chr_2010.reset_index(drop=True, inplace=True)
chr_2011.reset_index(drop=True, inplace=True)
chr_2012.reset_index(drop=True, inplace=True)
chr_2013.reset_index(drop=True, inplace=True)
chr_2014.reset_index(drop=True, inplace=True)
chr_2015.reset_index(drop=True, inplace=True)
chr_2016.reset_index(drop=True, inplace=True)
chr_2017.reset_index(drop=True, inplace=True)
chr_2018.reset_index(drop=True, inplace=True)
chr_2019.reset_index(drop=True, inplace=True)
chr_2020.reset_index(drop=True, inplace=True)
chr_2021.reset_index(drop=True, inplace=True)
chr_2022.reset_index(drop=True, inplace=True)
chr_2023.reset_index(drop=True, inplace=True)
print("Index Reset")

# Set all header text to lowercase
chr_2010.columns = chr_2010.columns.str.lower()
chr_2011.columns = chr_2011.columns.str.lower()
chr_2012.columns = chr_2012.columns.str.lower()
chr_2013.columns = chr_2013.columns.str.lower()
chr_2014.columns = chr_2014.columns.str.lower()
chr_2015.columns = chr_2015.columns.str.lower()
chr_2016.columns = chr_2016.columns.str.lower()
chr_2017.columns = chr_2017.columns.str.lower()
chr_2018.columns = chr_2018.columns.str.lower()
chr_2019.columns = chr_2019.columns.str.lower()
chr_2020.columns = chr_2020.columns.str.lower()
chr_2021.columns = chr_2021.columns.str.lower()
chr_2022.columns = chr_2022.columns.str.lower()
chr_2023.columns = chr_2023.columns.str.lower()
print("Headers Set to Lowercase")
```
